```python
# 모바일 게임 랭킹 데이터 불러오기
df = pd.read_csv('C:/Users/X1Carbon/Desktop/data_project/gameData/game_tree/mobile_all.csv')
df
```

```python
# 모바일 게임 랭킹 데이터 20개씩 자르기
i = df[df['rank']>20].index
mobile_top20 = df.drop(i)
mobile_top20.head(30)
```

```python
# 인덱스가 이상해서 인덱스 리셋 시켜주기
mobile_top20 = mobile_top20.reset_index()
mobile_top20
```

```python
# 컬럼명 수정
mobile_top20 = mobile_top20[['level_0', 'rank', 'change', 'name', 'service', 'genre']]
mobile_top20
```

```python
# week_num 추가
mobile_top20.insert(0, 'week_num', 0)
mobile_top20
```

```python
# week_num 순서 정렬
share = 185 - mobile_top20['level_0']//20
mobile_top20['week_num'] =share
```
