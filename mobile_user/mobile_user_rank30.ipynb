### csv 파일 불러와 30개 단위로 절삭 후 rank_30으로 저장

```python
folders = os.listdir('C:/Users/X1Carbon/Documents/dev/gameData/mobile_user')
folders
```

```python
df_all = pd.DataFrame()
for i in range(0,len(folders)):
    if folders[i].split('.')[1] == 'csv':
        file = 'C:/Users/X1Carbon/Desktop/mobile_user/' + folders[i] #슬레시 위치 정확하게 할 것!!
        df = pd.read_csv(file,encoding='utf-8')
        df = df[:30]
        df_all = pd.concat([df_all, df])

df_all
```

```python
df_all.to_csv("mobile_user_rank_30.csv", index = False, mode='w')
```
