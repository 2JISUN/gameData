# df_all : n개의 csv파일 합쳐서 불러오기 

```python
# 파일 리스트 불러오기
forders = os.listdir('C:/Users/snm03/Documents/debbu/gameData/the_log_csv')
print(forders)
```

```python
df_all = pd.DataFrame()
for i in range(0,len(forders)):
    if forders[i].split('.')[1] == 'csv':
        file = 'C:/Users/snm03/Documents/debbu/gameData/the_log_csv/'+forders[i] #슬레시 위치 정확하게 할 것!!
        df= pd.read_csv(file,encoding='utf-8') 
        df_all = pd.concat([df_all, df])
```
